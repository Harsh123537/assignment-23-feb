```python
'''In Pandas, a DataFrame is a two-dimensional labeled data structure, whereas a Series is a one-dimensional labeled array. A DataFrame is similar to a spreadsheet or a SQL table, with rows and columns.
Each column in a DataFrame can have a different data type (integer, float, string, etc.).

On the other hand, a Series is like a column in a DataFrame. It can be created from a list, ndarray or dictionary, and it has a label, which is used to index the values.'''

import pandas as pd

# creating a DataFrame
data = {'name': ['John', 'Sarah', 'Adam', 'Jess'], 
        'age': [25, 33, 19, 42], 
        'city': ['New York', 'London', 'Paris', 'Tokyo']}
df = pd.DataFrame(data)

# outputting the DataFrame
print(df)
print()

# creating a Series
s = pd.Series([25, 33, 19, 42], name='age')

# outputting the Series
print(s)
```

```python
#1)Drop

data={'Name':['harsh','mohan','ram','suresh'],'Age':[34,56,23,78],'Gender':['F','M','M','F']}
df=pd.DataFrame(data)
df1=df.drop('Gender',axis=1,inplace=False)
df1

#2)sort_values

df.sort_values(by='Age',ascending=False)

#3)groupby

df.groupby('Gender')['Age'].mean()
```

```python
import pandas as pd

# create two series
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series(['a', 'b', 'c', 'd', 'e'])

# create a DataFrame using the two series
df = pd.DataFrame({'col1': s1, 'col2': s2})

print(df)
```
